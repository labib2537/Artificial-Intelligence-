```
from termcolor import colored
data = []
x = 'random'

def line():
  print("-------------------")

def show_menu():
  line()
  print(colored("MENU",'green'))
  line()
  print("1. Add your course")
  print("2. View All course of the list")
  print("3. Remove or Unenroll from any courses")
  print("4. Exit")
  line()
  x = input("Enter Your choice: ")
  return x

while x!='4':
  x = show_menu()
  if x == '1': 
    course = input("Enroll your course: ")
    data.append(course) 
    print(course,'Course is added')
  elif x =='2':
    print("See your Enroll Courses:")
    for course in data:
      print(colored((course),'blue'))
  elif x =='3':
    course = input('Which course you wanna unenroll: ')
    if course in data:
      data.remove(course)
      print(course,'is unenrolled')
    else:
      print("You are not enroll any course") 
  elif x=='4':
      print(colored("Logged Out",'green'))
  else:
    print("WRONG CHOICE")
    print("Press 1,2,3 and 4")    


```
